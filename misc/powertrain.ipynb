```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 18}) # Set global font size to 18

dataset_path = "../data/powertrain-nk523"

corrente_tempo = pd.read_csv(f'{dataset_path}/corrente-tempo.csv', sep=";")
velocidade_tempo = pd.read_csv(f'{dataset_path}/velocidade-tempo.csv', sep=";")
velocidade_tempo
```

```python
plt.figure(figsize=(10,6))
plt.plot(corrente_tempo['tempo'], corrente_tempo['corrente'], label="Corrente (A)", color="red")

# Add labels and a title
plt.xlabel("t (s)")
plt.ylabel("Corrente (A)")

# Display the plot
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(velocidade_tempo['tempo'], velocidade_tempo['velocidade'], label="Velocidade (km/h)", color="blue")

# Add labels and a title
plt.xlabel("t (s)")
plt.ylabel("Velocidade (km/h)")

# Display the plot
plt.grid()
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,6)) # Create a figure and a set of subplots
ax1.plot(velocidade_tempo['tempo'], velocidade_tempo['velocidade'], color='blue', label='Velocidade (km/h)')
ax1.set_xlabel('Tempo (s)')
ax1.set_ylabel('Velocidade (km/h)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()

ax2.plot(corrente_tempo['tempo'], corrente_tempo['corrente'], color='red', label='Corrente (A)')
ax2.set_ylabel('Corrente (A)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout() # Adjust layout to prevent labels from overlapping
# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.show()
```

```python
# histograma da corrente (carga)
plt.figure(figsize=(10,6))
plt.hist(corrente_tempo['corrente'], label="Corrente (A)", color="red", bins=30, edgecolor="black")

# Add labels and a title
plt.title("DistribuiÃ§Ã£o da corrente (A) - Carga")
plt.ylabel("FrequÃªncia")
plt.xlabel("Corrente (A)")

# Display the plot
plt.grid()
plt.legend()
plt.show()
```
