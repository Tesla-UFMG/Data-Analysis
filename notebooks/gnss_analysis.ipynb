```python
import pandas as pd
import folium
```

```python
# Load CSV
df = pd.read_csv("../data/dados_gnss/volta_eng_130126.txt")
df.head(5)
```

```python
df = df[df["latitude"] != 0]
df = df[df["longitude"] != 0]
```

```python
# Create map centered on mean location
m = folium.Map(
    location=[df["latitude"].mean(), df["longitude"].mean()],
    zoom_start=12
)

# Plot each row as a point
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=3,
        color="blue",
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

# Save map
m.save("map.html")
```
