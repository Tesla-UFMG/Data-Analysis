```python
import pandas as pd
import folium
```

```python
# Load CSV
# dataset = "../data/dados_gnss/volta_eng_130126.txt"
dataset = "../data/dados_gnss/quarto_160126.txt"
dataset = "../data/dados_gnss/placa_2026/teste_andando_lucas_200226.txt"
dataset = "../data/dados_gnss/placa_2026/140426_estacionamento.txt"
df = pd.read_csv(dataset)
df.head(5)
```

```python
df = df[df["latitude"] != 0]
df = df[df["longitude"] != 0]
```

```python
# Create map centered on mean location
m = folium.Map(
    location=[df["latitude"].mean(), df["longitude"].mean()],
    zoom_start=12
)

# Plot each row as a point
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=3,
        color="blue",
        fill=True,
        fill_opacity=0.7
    ).add_to(m)

# Save map
m.save("map_140426.html")
```

```python

def nmea_checksum(sentence):
    cs = 0
    for c in sentence:
        cs ^= ord(c)
    return f"{cs:02X}"
    
cmd = "PSTMSETPAR,3102,0xA"
print(nmea_checksum(cmd)) # 0x12
```
